### ĐỒ ÁN MẠNG XÃ HỘI


Import thư viện

```python
# Network library
import networkx as nx
# Data libraries
import pandas as pd
import numpy as np
import numpy
# Plotting libraries
import matplotlib.pyplot as plt
from pyvis import network as net
import plotly.graph_objects as go
from pandas.core.frame import DataFrame
```

Load dữ liệu Book Dataset lên Project 

```python
data = pd.read_csv('./File Excel/Book_Dataset_1.csv');
data
```

```python
df=data.loc[:,['Category','Avilability']]
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Kiểm tra dữ liệu Null
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
from pandas.core.frame import DataFrame
df_book = pd.DataFrame()
df_book[['FromNodeID','ToNodeID']]=df[['Category','Avilability']]
df_book
```

```python
#Kiểm tra số nodes và edges của dataset
from networkx.algorithms import bipartite
Book_Graph = nx.Graph()
FromNodeID = df_book['FromNodeID']
ToNodeID = df_book['ToNodeID']
print("So the loai",FromNodeID.nunique())
print("So luong trong kho",ToNodeID.nunique())
print("So canh",len(df))
```

```python
for index, row in df_book.iterrows():
    Book_Graph.add_edge(row['FromNodeID'],row['ToNodeID'],weight=1)
Book_Graph.add_nodes_from(FromNodeID,bipartite=0)
Book_Graph.add_nodes_from(ToNodeID ,bipartite=1)
```

```python
plt.figure(figsize=(12,12))
pos = nx.spring_layout(Book_Graph)
fig, ax = plt.subplots(1,1, figsize=(8,8), dpi=150)
nx.draw_networkx(Book_Graph, pos = nx.drawing.layout.bipartite_layout(Book_Graph,ToNodeID),font_size=8,width=0.4)
```

```python
from networkx.algorithms import bipartite
G = bipartite.weighted_projected_graph(Book_Graph, ToNodeID)
plt.figure(figsize=(12, 12))

layout = nx.spring_layout(G)

nx.draw_networkx_nodes(G,layout, nodelist=ToNodeID,node_size=250, node_color='blue')
nx.draw_networkx_edges(G, layout, edge_color="#cccccc")
node_labels = dict(zip(ToNodeID, ToNodeID))
nx.draw_networkx_labels(G, layout, labels=node_labels)
plt.axis('off')

plt.title("Graph test")

plt. show()
```
